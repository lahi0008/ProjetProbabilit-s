```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
from random import*
import random,math
simul=5000
n=10
#1. Grâce au logiciel Python, on tire des réalisations ty1, ..., ynu des n variables tY1,    , Ynu
#2. On calcule une réalisation de la moyenne empirique Y n
#3. On répète 5 000 fois les étapes 1 et 2. On obtient ainsi 5 000 réalisations de la variable Y n.
for k in range(simul):
    S=np.cumsum(np.random.randint(1,n+1))/np.random.rand(n)

R=S/simul
M=max(np.abs(R))
x=np.linspace(-M,M,10)
print(R)
plt.hist(R,bins=x,histtype="step")
```

```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
from random import*
import random,math
simul=5000
n=10
for k in range(simul):
    S=np.cumsum(np.random.randint(1,n+1)/np.random.rand(n))
#À partir de la réalisation de la moyenne empirique ¯ yn, on calcule une réalisation
#de cette variable transformée comme Zn=sqrt(n)*(S-2/2)
Z=np.sqrt(n)*(S-2/2)
R=Z/simul
M=max(np.abs(R))
x=np.linspace(-M,M,1000)
print(R)
plt.hist(R,bins=x,normed=1,histtype="step",label="Histogramme")
plt.legend(loc="best")
```

```python
import matplotlib.pyplot as plt
import scipy.stats as sps
import numpy as np
n,m,sigma=10,2,4
X=2*(np.random.rand(m,n))
S=np.sum(X,axis=1)/(np.sqrt(n)*(sigma))
M=max(np.abs(S))
x=np.linspace(-M,M,1000)
y=sps.norm.pdf(x)
plt.plot(x,y,"r",label="densite")

plt.hist(S,bins=x,normed=1,histtype="step",label="Histogramme")
plt.legend(loc="best")
plt.title("TCL")
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import string
import urllib.request
import pylab
#Importation et affichage du livre
x = urllib.request.urlopen("https://r-stat-sc-donnees.github.io/LesMiserables1.txt")
livre=str(x.read(), encoding = "UTF-8")
print(livre)
```

```python
#Suppression des ponctuation
points=livre.strip(string.punctuation)
print(points)
```

```python
#Transformation majuscules / minuscules
change=points.lower()
change
```

```python
#Division du texte en jetons individuels

liste =livre.split()
liste_mots = [mots for mots in liste if type(mots) !='int'] 
liste_mots
```

```python
#Pour les mots (un, et, le, il, est) on calcule leurs nombres respectifs au fur et à mesure que le livre progresse

mots = ["un", "et", "le", "il", "est"] 
values, counts = np.unique(liste_mots, return_counts = True) 

nbre_mots = len(liste_mots)
frequence= []
for i, j in zip(values[counts.argsort()], counts[counts.argsort()]):
    
   
    if i in mots:
        print(f'Le mot {i} apparait {j} fois dans le livre.')
        frequence.append((j/nbre_mots)*100)
```

```python
#Tracer les proportions nmot ris {i sur le document dans un seul graphique
fig = plt.figure()

x = [1,2,3,4,5]

h = frequence
w = 0.5 
BarName = ['est', 'un', 'le', 'et', 'il'] 
plt.bar(x, h, w)

plt.xlim(0,6)
plt.ylim(0,3)

plt.ylabel('fréquence d''apparution (%)')
plt.xlabel('Les mots')
plt.title('Diagramme en barre des proportions de chaque mots ')

pylab.xticks(x, BarName, rotation=30, c='red')

plt.show()
```

# #Pourquoi ne pouvons nous pas appliquer le théorème centrale (TCL) directement ?

on ne peut pas utilser le le théorème centrale limité car:
La somme de variables aléatoires identiques et indépendantes ne suit aucune loi et ne tend pas vers une variable aléatoire gaussienne dans notre cas.



# Que devons-nous faire sur le texte pour appliquer le TCL?


on doit augmenter la taille de notre échantillon car plus ce dernier est grand, plus on a de chances de vérifier que l'allure du diagramme en bâton ou de l'histogramme des échantillons se rapprochent fortement de la coube gaussienne.
